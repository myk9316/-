# Step 4. 모델 학습을 위한 데이터 전처리

```python
from sklearn.preprocessing import RobustScaler

# RobustScaler를 이용해 수치형 데이터를 표준화하기
X_num = data[['num_reviews','num_article', 'netizen_ratings']]

scaler = RobustScaler()
scaler.fit(X_num)
X_scaled = scaler.transform(X_num)
X_scaled = pd.DataFrame(X_scaled, index=X_num.index, columns=X_num.columns)

# get_dummies를 이용해 범주형 데이터를 one-hot 벡터로 변경하기
X_cat = data.drop(['num_viewers', 'num_reviews', 'num_article', 'netizen_ratings'], axis=1)
X_cat = pd.get_dummies(X_cat, drop_first=True)

# 입출력 데이터 통합하기
X = pd.concat([X_scaled, X_cat], axis=1)
y = data['num_viewers']
```

```python
X_cat
```

```python
X.shape
```

# Step 5. 모델 학습 및 평가

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
print(X_train.shape)
print(y_train.shape)
```

## Linear Regression 모델

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(X_train, y_train)

print('테스트 점수:', model.score(X_test, y_test))
```

## Ridge 모델

```python
from sklearn.linear_model import Ridge
model_ridge = Ridge()

model_ridge .fit(X_train, y_train)

print('테스트 점수:', model_ridge .score(X_test, y_test))
```

## Lasso 모델

```python
from sklearn.linear_model import Lasso
model_lasso = Lasso()

model_lasso .fit(X_train, y_train)


print('테스트 점수:', model_lasso.score(X_test, y_test))
```

## ElasticNet 모델

```python
from sklearn.linear_model import ElasticNet
model_Elastic= ElasticNet()

model_Elastic .fit(X_train, y_train)

print('테스트 점수:', model_Elastic.score(X_test, y_test))
```
